Implement a preorder traversal for directory listing.  
* Each directory name is stored in a TreeNode below.


```python
# TreeNode Definition
class TreeNode():
    def __init__(self, s: str, k: int):
        self.name = s
        self.ary = k
        self.child = [None]*k
```
