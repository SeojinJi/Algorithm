2326. Spiral Matrix IV
You are given two integers m and n, which represent the dimensions of a matrix.

You are also given the head of a linked list of integers.

Generate an m x n matrix that contains the integers in the linked list presented in spiral order (clockwise), starting from the top-left of the matrix. If there are remaining empty spaces, fill them with -1.

Return the generated matrix.


Input: m = 3, n = 5, head = [3,0,2,6,8,1,7,9,4,2,5,5,0]
Output: [[3,0,2,6,8],[5,0,-1,-1,1],[5,2,4,9,7]]
Explanation: The diagram above shows how the values are printed in the matrix.
Note that the remaining spaces in the matrix are filled with -1.

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
def create_linked_list(l: list) -> ListNode:
    if len(l) == 0:
        return None

    start = ListNode(l[0])
    node = start

    for i in range(1, len(l)):
        node_new = ListNode(l[i])
        node.next = node_new
        node = node_new

    return start
```

```python
class Solution:
    def spiralMatrix(self, m: int, n: int, head : ListNode) :
        total = m * n
        number = 1
        arr = []
        while head is not None :
            arr.append(head.val)
            head = head.next
            number += 1
        while number <= total :
            arr.append(-1)
            number += 1
        result = [[0] * n for _ in range(m)]
        num = 0
        top = left = 0
        bottom = m - 1
        right = n - 1

        while num < total :
            #위쪽 행
            for i in range(left, right+1) :
                result[top][i] = arr[num]
                num += 1
            top += 1
            if num >= total :
                break

            #오른쪽 열
            for i in range(top, bottom + 1) :
                result[i][right] = arr[num]
                num += 1
            right -= 1
            if num >= total :
                break

            #아래쪽 행
            for i in range(right, left -1, -1) :
                result[bottom][i] = arr[num]
                num += 1
            bottom -= 1
            if num >= total :
                break

            #왼쪽 열
            for i in range(bottom, top - 1, -1) :
                result[i][left] = arr[num]
                num += 1
            left += 1
            if num >= total :
                break

        return result

```

```python
head = [3,0,2,6,8,1,7,9,4,2,5,5,0]
m = 3
n = 5
listA = create_linked_list(head)
S = Solution()
S.spiralMatrix(m, n, listA)
```
