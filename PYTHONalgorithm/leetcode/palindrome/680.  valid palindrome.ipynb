680. Valid Palindrome II

Given a string s, return true if the s can be palindrome after deleting at most one character from it.

 

Example 1:

Input: s = "aba"
Output: true
Example 2:

Input: s = "abca"
Output: true
Explanation: You could delete the character 'c'.
Example 3:

Input: s = "abc"
Output: false
 

Constraints:

1 <= s.length <= 105
s consists of lowercase English letters.

```python
class Solution:
    def onedelete(self, s, i) :
        if i == len(s)-1 :
            result = s[:i]
        else : 
            result = s[:i] + s[i+1:]
        return result
    def validPalindrome(self, s: str) -> bool: # time limit exceed
        if s == s[::-1] :
            return True
        length = len(s)
        for i in range(length) :
            result = self.onedelete(s, i)
            if result == result[::-1] :
                return True
        return False
```

```python
class Solution:
    def validPalindrome(self, s: str) -> bool: # Solution
        p1 = 0
        p2 = len(s) - 1
        while p1 <= p2 :
            if s[p1] != s[p2] :
                result1 = s[:p1] + s[p1+1:]
                result2 = s[:p2] + s[p2+1:]
                return (result1 == result1[::-1]) or (result2 == result2[::-1])
            p1 += 1
            p2 -= 1
        return True
```

```python
s = ""
S = Solution()
S.validPalindrome(s)
```
